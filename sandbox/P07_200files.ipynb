```python
def cv_mq(pgs):
    sam = pgs[[_ for _ in pgs.columns if 'LFQ' in _]]
    sam = sam.replace(0, np.nan)
    cv = (sam.std(axis=1)) / (sam.mean(axis=1))
    vals = cv.sort_values()
    vals = vals[~vals.isna()]
    
    return vals


def cv_ap(x):
    
    lfq = [_ for _ in x.columns if 'LFQ' in _]
    cv = (x[lfq].std(axis=1)) / (x[lfq].mean(axis=1))

    vals = cv.sort_values()
    vals = vals[~vals.isna()]
    
    return vals
```

```python
v_mq = cv_mq(pgs)
v_ap = cv_ap(pgs)
```

```python
plt.plot(v_mq*100, np.arange(len(v_mq)), label='MQ')

plt.plot(v_ap*100, np.arange(len(v_ap)), label='AP')
plt.legend()

plt.xlim([0,25])
plt.show()
```
